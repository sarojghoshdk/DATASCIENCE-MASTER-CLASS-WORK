```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x , y)
```

```python
x
```

```python
y
```

```python
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x , y , c = colours , s = sizes , alpha=.5)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("random data")
```

```python
x = ["a", "b", "c", "d", "e"]   # 5 catagorical data
```

```python
y = np.random.rand(5)     # 5 random numerical data
plt.bar(x , y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("bar plot")
```

```python
y
```

```python
# bar plot shows in horizontally
plt.barh(x , y)
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("bar plot")
```

```python
plt.figure(figsize=(6,2))
plt.plot(x , y , '--b')
plt.xlabel("This is my x data")
plt.ylabel("This is my y data")
plt.title("This is my title")
plt.grid()
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000*np.random.rand(50)
plt.scatter(x , y , c = '#00FF00' , s = sizes , alpha=.5 , marker='v')
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.title("random data")
```

```python
# to show 3D data
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)

fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.show()
```
